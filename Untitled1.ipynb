$$
\vec{E}=E_0\hat{x}\cos(\omega t- kz) \\
\vec{B}=\frac{E_0}{c}\hat{y}\cos(\omega t- kz) \\
I=\left< \vec{S}\right> = \frac{E_0^2}{2\mu_0 c} \\
E_0 = \sqrt{2\mu_0 c I}
$$


```python
I = 1300.0
e = -1.6021766*10**(-19)
m = 1.67262*10**(-24)
mu0 = 1.257*10**(-6)
c=3.0*10**8
E0 = sqrt(2*mu0*c*I)
```

Por lo visto en clase, para un rayo de luz monocromático

$$x(t) = \frac{qE_0}{\omega^2m}(1-\cos(\omega t)) \\
z(t) = \frac{1}{8}\frac{\omega}{c}\left(\frac{qE_0}{\omega^2m}\right)^2(2\omega t - \sin(2\omega t))$$

Es necesario recalcar que esta solución se halló haciendo una perturbación de primer orden en $\vec B$ de la solución hallada solo para un campo eléctrico, Esto es una buena aproximación ya que $B \propto E_0/c$.

Una longitud de onda típica para la luz proveniente del sol es $\lambda = 516nm$

```python
l = 516.0*10**(-9)
w= 2*pi*c/l
tau = 1.0/w
```

```python
t = linspace(0,100*tau,1000)
x = e*E0/(w**2.0*m)*(1.0-cos(w*t))
y = zeros(len(t))
z = 1.0/8.0*w/c*(e*E0/(w**2.0*m))**2*(2*w*t-sin(2*w*t))
```

```python
plot(x,z)
plt.savefig('trayectoria2d.pdf', format = 'pdf', transparent = True)
```

```python
from mpl_toolkits.mplot3d import axes3d
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot(x,y,z)
plt.savefig('trayectoria3d.pdf', format = 'pdf', transparent = True)
```
