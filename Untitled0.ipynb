```python
x = linspace(-3.0,3.0,100)
y = linspace(-3.0,3.0,100)
```

```python
def crossk(a):
    new = zeros(3)
    new[0] = a[1]
    new[1] = -a[0]
    new[2] = 0
```

```python
Sx = zeros((100,100))
Sy = zeros((100,100))
for i in range(100):
    for j in range(100):
        Sx[i,j] = y[i]/(sqrt((x[i]-1)**2+y[i]**2))**3 - y[i]/(sqrt((x[i]+1)**2+y[i]**2))**3
        Sy[i,j] = -(x[i]-1)/(sqrt((x[i]-1)**2+y[i]**2))**3 + (x[i]+1)/(sqrt((x[i]+1)**2+y[i]**2))**3
magS = sqrt(Sx**2+Sy**2)
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,8))

ax.set_xlabel("$x$", size ='16')
ax.set_ylabel("$y$", size ='16')
ax.set_title(r"$\vec E$", size ='16')
ax.streamplot(linspace(0,len(x)-1,len(x)),linspace(0,len(y)-1,len(y)),Sx,Sy, color ='g')
im = ax.imshow(magS)
fig.colorbar(im, ax=ax)

figurename="S"
savefig(figurename + '.pdf',format = 'pdf', transparent=True)
```
