```python
a=3*10**(-2)
w=2*pi
c=3*10**8
y=linspace(-10**(10),10**(10),1000) #r se puede tomar negativo para que aparezca todo el plano, y E solo depende del cuadrado de r
z=linspace(-10**(10),10**(10),1000)
```

```python
def E_f(y,z,t):
    return abs(a*w**2/c**2*(abs(y)/(y**2+z**2))*cos(w*(t-sqrt(y**2+z**2)/c)))
```

```python
E=np.zeros((len(y),len(z)))
for i in range(len(y)):
    E[i,:]=E_f(y[i],z,0)
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,8))

ax.set_xlabel("$z[m]$", size ='16')
ax.set_ylabel("$y[m]$", size ='16')
ax.set_title("$E(t=0)$", size ='16')
im = ax.imshow(E,extent=[-10**10,10**10,-10**10,10**10],vmin=0,vmax=10**(-26)) #Se pone un máximo para poder verlo bien, evitando que solo se vea un punto rojo en donde está la partícula
fig.colorbar(im, ax=ax)

figurename="E_plot_0"
savefig(figurename + '.pdf',format = 'pdf', transparent=True)
```

```python
E=np.zeros((len(y),len(z)))
for i in range(len(y)):
    E[i,:]=E_f(y[i],z,0.25)
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,8))

ax.set_xlabel("$z[m]$", size ='16')
ax.set_ylabel("$y[m]$", size ='16')
ax.set_title("$E(t={2\pi}/{4\omega})$", size ='16')
im = ax.imshow(E,extent=[-10**10,10**10,-10**10,10**10], vmin=0,vmax = 10**(-26))
fig.colorbar(im, ax=ax)

figurename="E_plot_0.25"
savefig(figurename + '.pdf',format = 'pdf', transparent=True)
```
