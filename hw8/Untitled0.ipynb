#Campo electromagnético de una partícula cargada con velocidad constante
Los campos hallados durante la tarea anterior fueron:
$$
\phi(\vec r,t)=\frac{\gamma q}{4\pi\epsilon_0}\frac{1}{\sqrt{x^2+y^2+\gamma^2(z-vt)^2}}\\
\vec A(\vec r,t)=\frac{\mu_0\gamma q v \hat k}{4\pi}\frac{1}{\sqrt{x^2+y^2+\gamma^2(z-vt)^2}} \\
\vec E(\vec r,t)=\frac{\gamma q}{4\pi\epsilon_0}\frac{x\hat x+y\hat y+(z-vt)\hat z}{(x^2+y^2+\gamma^2(z-vt)^2)^{3/2}}\\
\vec B(\vec r,t)=\frac{\mu_0\gamma q v}{4\pi}\frac{-y\hat x+x\hat y}{(x^2+y^2+\gamma^2(z-vt)^2)^{3/2}}
$$
Tomando la norma de $\vec E$ y $\vec B$ sobre el plano $x=0$
$$
E=\frac{\gamma q}{4\pi\epsilon_0}\frac{\sqrt{y^2+(z-vt)^2}}{(y^2+\gamma^2(z-vt)^2)^{3/2}} \\
B=\frac{\mu_0\gamma q v}{4\pi}\frac{y^2}{(y^2+\gamma^2(z-vt)^2)^{3/2}}
$$
Abajo se puede modificar el valor de $\beta = v/c$

```python
beta = 0.9
```

```python
gamma = 1/sqrt(1-beta**2)
y=linspace(-1.0,1.0,1000) #r se puede tomar negativo para que aparezca todo el plano, y E solo depende del cuadrado de r
z=linspace(-1.0,1.0,1000)
```

```python
def E_f(y,z,t,g):
    return g*sqrt(y**2+(z-beta*t)**2)/(y**2+g**2*(z-beta*t)**2)
```

```python
E=np.zeros((len(y),len(z)))
for i in range(len(y)):
    E[i,:]=E_f(y[i],z,0,gamma)
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,8))

ax.set_xlabel("$z[m]$", size ='16')
ax.set_ylabel("$y[m]$", size ='16')
ax.set_title("$E(t=0)$", size ='16')
im = ax.imshow(E,extent=[-1,1,-1,1], vmin=0, vmax=50) #Se pone un máximo para poder verlo bien, evitando que solo se vea un punto rojo en donde está la partícula
fig.colorbar(im, ax=ax)

figurename="E_plot_"+str(beta)
savefig(figurename + '.pdf',format = 'pdf', transparent=True)
```
