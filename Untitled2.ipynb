#Radiación partícula puntual

##Velocidad Paralela a acelaración
$$
\frac{dP}{d\Omega} = \frac{\mu_0q^2a^2}{16\pi^2c}\frac{\sin^2\theta}{(1-\beta \cos\theta)^5}
$$

```python
theta = linspace(0,2*pi,1000)
b = 0.9
```

```python
def P(t,b):
    return sin(t)**2/(1-b*cos(t))**5
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.plot(P(theta,b)*cos(theta), P(theta,b)*sin(theta),label=str(b))
ax.scatter(0,0, alpha =1, s=100)
```

##Velocidad perpendicular a aceleración
$$
\frac{dP}{d\Omega} = \frac{e^2a^2}{16\pi^2\epsilon_0c}\frac{1}{(1-\beta \cos\theta)^5}\left[(1-\beta \cos\theta)^2-\frac{1}{\gamma^2}\sin^2\theta\cos^2\phi\right]
$$

```python
def P2(t,p,b):
    return 1/(1-b*cos(t))**5*((1-b*cos(t))**2-(1-b**2)*sin(t)**2*cos(p)**2)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].plot(P2(theta,0,b)*sin(theta), P2(theta,0,b)*cos(theta),label=str(b))
ax[0].scatter(0,0, alpha =1, s=100)
ax[0].set_xlabel("$x$",size=16)
ax[0].set_ylabel("$z$",size=16)

ax[1].plot(P2(theta,pi/2,b)*sin(theta), P2(theta,pi/2,b)*cos(theta),label=str(b))
ax[1].scatter(0,0, alpha =1, s=100)
ax[1].set_xlabel("$y$",size=16)
ax[1].set_ylabel("$z$",size=16)

ax[2].plot(P2(pi/2,theta,b)*cos(theta), P2(pi/2,theta,b)*sin(theta),label=str(b))
ax[2].scatter(0,0, alpha =1, s=100)
ax[2].set_xlabel("$x$",size=16)
ax[2].set_ylabel("$y$",size=16)
```
